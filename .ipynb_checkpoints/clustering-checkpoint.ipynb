# Problem 3: Clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import warnings

# Suppress warnings
warnings.filterwarnings('ignore')

# Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
column_names = [
    'Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 
    'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 
    'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'
]
data = pd.read_csv(url, header=None, names=column_names)

# Preprocess the data
data.drop('Class', axis=1, inplace=True)  # Remove the class column as it's not part of the clustering
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Determine the optimal number of clusters using the Yellowbrick's Elbow method
model = KMeans(random_state=0)
visualizer = KElbowVisualizer(model, k=(1,11))
visualizer.fit(data_scaled)  # Fit the data to the visualizer
visualizer.show()  # Finalize and render the figure

# Use the optimal number from the elbow plot
optimal_k = visualizer.elbow_value_
print(f"Optimal number of clusters: {optimal_k}")

# Clustering with K-means using the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
clusters_kmeans = kmeans.fit_predict(data_scaled)

# Clustering with DBSCAN
# Use NearestNeighbors to find the optimal eps
neighbors = NearestNeighbors(n_neighbors=2)
neighbors_fit = neighbors.fit(data_scaled)
distances, indices = neighbors_fit.kneighbors(data_scaled)

# Sort and plot distances
distances = np.sort(distances, axis=0)
distances = distances[:, 1]
plt.plot(distances)
plt.title('K-Distance Graph')
plt.xlabel('Points sorted by distance')
plt.ylabel('Epsilon distance')
plt.show()

eps_value = 2.5

dbscan = DBSCAN(eps=eps_value, min_samples=5)
clusters_dbscan = dbscan.fit_predict(data_scaled)


# Analyze and visualize the results from K-means
data['KMeans_Cluster'] = clusters_kmeans
print(data.groupby('KMeans_Cluster').mean())  # Print the mean values of features per cluster

# Visualization of clusters for K-means
plt.figure(figsize=(10, 6))
plt.scatter(data['Alcohol'], data['Malic acid'], c=data['KMeans_Cluster'], cmap='viridis', marker='o')
plt.title('K-Means Cluster Visualization')
plt.xlabel('Alcohol')
plt.ylabel('Malic acid')
plt.colorbar(label='Cluster Label')
plt.show()

# Visualization of clusters for DBSCAN
plt.figure(figsize=(10, 6))
plt.scatter(data['Alcohol'], data['Malic acid'], c=clusters_dbscan, cmap='viridis', marker='o')
plt.title('DBSCAN Cluster Visualization')
plt.xlabel('Alcohol')
plt.ylabel('Malic acid')
plt.colorbar(label='Cluster Label')
plt.show()
```

# Analysis of Results

### Elbow Graph Interpretation
- The Elbow graph for K-Means suggests that the **optimal number of clusters is 3**, as indicated by the elbow at \( k = 3 \). This point represents a balance between minimizing the within-cluster sum of squares (distortion score) and the complexity of the model (increasing number of clusters).

### Cluster Profiles (K-Means)
- The cluster mean values for different chemical properties like Alcohol, Malic Acid, and others show distinct profiles for each cluster:
  - **Cluster 0**: This cluster seems to characterize wines with moderate alcohol content but higher levels of nonflavanoid phenols compared to the other clusters.
  - **Cluster 1**: Wines in this cluster tend to have the lowest average alcohol content and ash but moderate levels of flavanoids.
  - **Cluster 2**: This cluster includes wines with the highest average alcohol content and magnesium levels, suggesting these might be richer or more robust wines.

### Cluster Visualizations (K-Means)
- The visualization shows clear separation among the three clusters based on Alcohol and Malic acid. The clusters seem well-separated, suggesting good clustering performance by K-Means on this dataset.

## DBSCAN Clustering Analysis

### Parameter Setup and Performance
- The DBSCAN algorithm was adjusted to better fit the dataset with an `eps` value suggested by the K-Distance graph. Despite adjustments, the visualization indicates a significant number of data points classified as noise (shown in purple), with fewer distinct clusters identified compared to K-Means.

### Cluster Visualization (DBSCAN)
- Unlike K-Means, DBSCAN results show a less clear clustering pattern, with many points failing to form clusters. This suggests that DBSCAN may not be as effective for this particular dataset due to its density-based clustering nature, which may require more fine-tuning of parameters or may not suit the distribution of data points as well as K-Means.

## Summary
- **K-Means** provides a quick and simple way to partition a dataset and works best when clusters are spherical. It is most effective when the number of clusters and the distribution of cluster size are known or can be estimated. With this dataset, it provides clear and distinct clustering, making it suitable for other datasets like this where groups are well-separated based on their chemical properties.
- **DBSCAN** excels in identifying clusters of arbitrary shapes and sizes, making it suitable for applications where the cluster model isn't well-defined. It is particularly useful when dealing with spatial data and noise. Here, it shows challenges in identifying meaningful clusters without extensive parameter tuning, highlighting its sensitivity to parameter settings and the distribution of data points.
